### Q1. Write a Python function to calculate total spend per customer

````python
# Raw transaction logs
transactions_data = [
    {"customer": "Amit", "spend": 1200},
    {"customer": "Siddhant", "spend": 500},
    {"customer": "Priya", "spend": 800},
    {"customer": "Amit", "spend": 300},
    {"customer": "Siddhant", "spend": 1500},
]

transactions_df = pd.DataFrame(transactions_data)
display(transactions_df)

# result_df = transactions_df.groupby("customer", as_index=False).sum()
result_df = transactions_df.groupby("customer", as_index=False).agg({"spend": "sum"})
display(result_df)
````

````python
def calculate_total_spend(transactions):
    # Initialize a dictionary that defaults missing keys to an integer (0)
    spend_summary = defaultdict(int)
    
    # Loop through the list and aggregate values
    for row in transactions:
        customer = row["customer"]
        spend = row["spend"]
        spend_summary[customer] += spend
        
    # Convert back to a standard dictionary for the final output
    return dict(spend_summary)

# Executing the function
print(calculate_total_spend(transactions_data))
# Output: {'Amit': 1500, 'Siddhant': 2000, 'Priya': 800}
````

#### Great use of Pandas. Now, how would you solve this using `pure vanilla Python` without importing any external libraries like Pandas or software packages?

Let's break down the classic, high-performance way to do this in raw Python so you can add it to your toolkit.

---

##### 🐍 The Pure Python Approach (Using `defaultdict`)

When you don't have Pandas available (like in a lightweight AWS Lambda function or a basic Python script), the most efficient way to aggregate data is by using a dictionary. Specifically, `defaultdict` from Python's built-in `collections` module makes this incredibly elegant.

Here is how you write it:

```python
from collections import defaultdict

def calculate_total_spend(transactions):
    # Initialize a dictionary that defaults missing keys to an integer (0)
    spend_summary = defaultdict(int)
    
    # Loop through the list and aggregate values
    for row in transactions:
        customer = row["customer"]
        spend = row["spend"]
        spend_summary[customer] += spend
        
    # Convert back to a standard dictionary for the final output
    return dict(spend_summary)

# Executing the function
print(calculate_total_spend(transactions_data))
# Output: {'Amit': 1500, 'Siddhant': 2000, 'Priya': 800}

```

---

##### 🛠️ Performance Analysis: Pandas vs. Pure Python

Understanding *when* to use which tool is a critical trait of a Data Engineer.

| Metric | Pure Python (`defaultdict`) | Pandas (`pd.DataFrame`) |
| --- | --- | --- |
| **Time Complexity** | O(N) — Loops through the list exactly once. Highly efficient. | O(N) — High-speed C-optimized vectors under the hood. |
| **Memory Overhead** | **Very Low** — Creates a single lightweight dictionary structure. | **High** — Wrapping lists into DataFrames creates massive metadata overhead. |
| **Best Used For...** | Small to medium payloads, API responses, configuration scripts, and serverless environments. | Large tabular datasets, complex feature engineering, and matrix operations. |

### Q2. Write Python logic to detect consecutive job failures.

#### 📋 The Concept: What is a "Consecutive Failure"?

Imagine an automated data orchestration pipeline (like Azure Data Factory or Apache Airflow) running a specific data engineering job every hour. The system generates a log status of either `"SUCCESS"` or `"FAILURE"`.

The monitoring team wants an alert triggered **only if the job fails N times in a row**.

* If a job fails, passes, and fails again—that is **not** consecutive. The streak breaks.
* If a job fails 3 times back-to-back, that **is** a consecutive failure streak of 3.

#### 🎯 Expected Input & Output Examples
<br>

```
Let's say the team sets the threshold to **3 consecutive failures** to trigger an alert.
Example Scenario A:
* **Input Log List:** `["SUCCESS", "FAILURE", "FAILURE", "SUCCESS", "FAILURE", "FAILURE"]`
* **Threshold:** `3`
* **Expected Output:** `False` (The failure streak reached a maximum of 2 before a `"SUCCESS"` wiped it out).
Example Scenario B:
* **Input Log List:** `["SUCCESS", "FAILURE", "FAILURE", "FAILURE", "SUCCESS", "SUCCESS"]`
* **Threshold:** `3`
* **Expected Output:** `True` (Indices 1, 2, and 3 represent 3 failures in a row).
```

---

````python
def check_consecutive_failures(log_list, threshold):
    # Check if the number of 'FAILURE' logs reaches the threshold
    threshold_reached = threshold
    for log in log_list:
        if log == "FAILURE":
            threshold_reached -= 1
        else:
            threshold_reached = threshold
        if threshold_reached == 0:
            print("Consecutive Failure:-",True)
            break
    else:   
        # This else block executes if the for loop completes without hitting 'break'
        print("Consecutive Failure:-",False)

# Additional test cases
check_consecutive_failures(["FAILURE", "SUCCESS", "FAILURE"], 2)
check_consecutive_failures(["SUCCESS", "SUCCESS", "SUCCESS"], 1)
check_consecutive_failures(["FAILURE", "SUCCESS", "FAILURE", "FAILURE", "FAILURE"], 3)
check_consecutive_failures(["FAILURE", "FAILURE", "SUCCESS", "FAILURE", "FAILURE", "FAILURE"], 2)
check_consecutive_failures(["FAILURE", "FAILURE", "SUCCESS", "FAILURE", "FAILURE", "FAILURE"], 4)
````
